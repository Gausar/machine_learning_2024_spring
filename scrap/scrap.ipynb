```python
import requests
import urllib.request
from bs4 import BeautifulSoup
```

```python
baseurl = 'https://news.num.edu.mn/?cat=31&paged='
news_link_list = []
for i in range(1, 2):
    #1 ees 495 r huudas hurtel for yvuulna
    url = baseurl + str(i)
    print(url)
    response = requests.get(url)
    if response.status_code != 200:
        print(response.status_code)
        print('error ', url)
        continue
    #print(response.text)
    soup = BeautifulSoup(response.text, "html.parser")
    li_list = soup.find_all("div", {"class": "media"})
    for li in li_list:
        a = li.find('a')
        news_url = a['href']
        print(news_url)
        news_link_list.append(news_url)
```
